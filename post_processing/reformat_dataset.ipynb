```python
import csv

input_file = 'test.csv'
input_encoding = 'utf-8'

output_file_1 = 'test1.csv'
output_file_2 = 'test2.csv'

with open(input_file, 'r', newline='', encoding=input_encoding) as infile:
    reader = csv.reader(infile)
    header = next(reader)  
    with open(output_file_1, 'w', newline='', encoding=input_encoding) as outfile_1:
        with open(output_file_2, 'w', newline='', encoding=input_encoding) as outfile_2:
            writer_1 = csv.writer(outfile_1)
            writer_1.writerow(header)  # Write the header row
            i = 0
            for row in (reader):
                if i < 5:
                    writer_1.writerow(row)
                else:
                    writer_2 = csv.writer(outfile_2)
                    # writer_2.writerow(header)  # Write the header row
                    writer_2.writerows(reader)
                i += 1
```

```python
input_encoding = 'utf-8'
def concatenate_csv(input_file1, input_file2, output_file):
    with open(input_file1, 'r', newline='', encoding=input_encoding) as file1, \
         open(input_file2, 'r', newline='', encoding=input_encoding) as file2, \
         open(output_file, 'w', newline='', encoding=input_encoding) as output:
        
        reader1 = csv.reader(file1)
        reader2 = csv.reader(file2)
        writer = csv.writer(output)

        # Write headers
        headers = next(reader1)
        writer.writerow(headers)

        # Write data from first file
        for row in reader1:
            writer.writerow(row)

        next(reader2)

        # Write data from second file
        for row in reader2:
            writer.writerow(row)
```

```python

# Example usage
concatenate_csv('test1.csv', 'test2.csv', 'test3.csv')
```

```python
import csv

input_file = 'matches.csv'
input_encoding = 'utf-8'

output_file_1 = 'data1.csv'
output_file_2 = 'data2.csv'

with open(input_file, 'r', newline='', encoding=input_encoding) as infile:
    reader = csv.reader(infile)
    header = next(reader)  
    with open(output_file_1, 'w', newline='', encoding=input_encoding) as outfile_1:
        with open(output_file_2, 'w', newline='', encoding=input_encoding) as outfile_2:
            writer_1 = csv.writer(outfile_1)
            writer_1.writerow(header)  # Write the header row
            remaining_rows = []
            i = 0
            for row in reader:
                if i < 4560:
                    writer_1.writerow(row)
                else:
                    remaining_rows.append(row)
                i += 1
            writer_2 = csv.writer(outfile_2)
            writer_2.writerow(header)  # Write the header row
            writer_2.writerows(remaining_rows)
```
