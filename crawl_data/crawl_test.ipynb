```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
team_name = 'chelsea'
year = 2013
team_url = 'https://fbref.com/en/squads/cff3d9bb/2012-2013/Chelsea-Stats'
data = requests.get(team_url)
soup = BeautifulSoup(data.text)
```

```python
# Scores & Fixtures
print(f"*******Crawling Scores & Fixtures for {team_name} in season {year-1}-{year}")
try:
    matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
except ValueError:
    print(f"Could not find Scores & Fixtures for {team_name} in {year-1}-{year}")
# Shooting
soup = BeautifulSoup(data.text)
links = [l.get("href") for l in soup.find_all('a')]
links2 = [l for l in links if l and 'all_comps/shooting/' in l]

if links2:
    data2 = requests.get(f"https://fbref.com{links2[0]}")
    shooting = pd.read_html(data2.text, match="Shooting")[0]
    shooting.columns = shooting.columns.droplevel()
else:
    print(f"Could not find Shooting for {team_name} in {year-1}-{year}")
```
